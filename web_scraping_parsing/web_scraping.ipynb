```python
with open('euler_problems.html') as data:
    contents = data.read()
print(contents[:100])
```

```python
from bs4 import BeautifulSoup
broken_html = '<ul class=country><li>Area<li>Population</ul>'
soup = BeautifulSoup(broken_html, 'html.parser')
fixed_html = soup.prettify()
print(fixed_html)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(contents,'html.parser')
table = soup.find('table', attrs={'id':'problems_table'})
#print(table.find('tr').prettify())
trs = table.find_all('tr')
for tr in trs:
    print(tr.prettify())
```

```python
for tr in trs:
    tds = tr.find_all('td')
    #print(tds)
    for td in tds:
        print(td.text)
```

```python
result={}
for tr in trs:
    tds = tr.find_all('td')
    if tds:
        #print(tds[0],tds[0].text)
        result[tds[0].text]={"title":tds[1].text, "solved":tds[2].text}
result
```
