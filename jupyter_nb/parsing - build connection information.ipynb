```python
from bs4 import BeautifulSoup
with open('work_test.htm') as file:
    data = file.read()
soup = BeautifulSoup(data,'html.parser')

main = soup.find('div', attrs={'id':'Page'})
len(main)
```

```python
divpm = main.findAll('div',attrs = {'parameters':True})
len(divpm)
```

```python
total = {}
for i,div in enumerate(divpm):
    
    total[div['id']] = {'id':i}
    
    param = {}
    for p in div['parameters'].split(';'):
        param.update(dict(zip(p.split(':')[::2], p.split(':')[1::2])))
    total[div['id']].update(param)
    
    style = {}
    for s in div['style'].split('; '):
        style.update(dict(zip(s.split(':')[::2], s.split(': ')[1::2])))
    total[div['id']].update(style)
    
    properties = {}
    for t in div['hdxproperties'].split(';'):
        properties.update(dict(zip(t.split(':')[::2], t.split(':')[1::2])))
    total[div['id']].update(properties)
    
total
```

```python
result = {}
for i in total:
    temp = {}
    temp['TOP'] = total[i]['TOP']
    temp['LEFT'] = total[i]['LEFT']
    temp['HEIGHT'] = total[i]['HEIGHT']
    temp['LEFWIDTHT'] = total[i]['WIDTH']
    if 'Rotation' in total[i]:
        temp['Rotation'] = total[i]['Rotation']
    result[i] = temp
result
```
