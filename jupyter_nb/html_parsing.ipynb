```python
#div > id & parameters 출력

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        if tag == 'div':
            for attr in attrs:
                if 'id' in attr:
                    print("1. Encountered an start tag :", tag)
                    print("     attr:", attr)
                if 'parameters' in attr:
                    print("   attr(param):", attr)
                    print("   =================================")                    


filename = 'MA_CH_OS1_WAXY.htm'
with open(filename) as file_object:
    contents = file_object.read()
    
parser = MyHTMLParser()
parser.feed(contents)
```

```python
#id & parameters 출력 -> List에 저장.

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    totalList=[]
    def handle_starttag(self, tag, attrs):
        if tag == 'div':
            subList=[]
            for attr in attrs:    
                if 'id' in attr:
                    print("1."+tag)
                    print("   attr(id)   :", attr)
                    subList=list(attr)
                    #print(subList)
                if 'parameters' in attr:
                    print("   attr(param):", attr)
                    subList+=list(attr)
                    #print(subList)
            #print("   =================================")
            self.totalList.append(subList)


filename = 'MA_CH_OS1_WAXY.htm'
with open(filename) as file_object:
    contents = file_object.read()
    
parser = MyHTMLParser()
parser.feed(contents)
#print(parser.totalList)
for i in range(len(parser.totalList)):
    print(parser.totalList[i])
```

```python
# List에 저장 > List 선택적 출력

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    totalList=[]
    def handle_starttag(self, tag, attrs):
        if tag == 'div':
            subList=[]
            for attr in attrs:    
                if 'id' in attr:
                    subList=list(attr)
                if 'parameters' in attr:
                    subList+=list(attr)
            self.totalList.append(subList)

filename = 'MA_CH_OS1_WAXY.htm'
with open(filename) as file_object:
    contents = file_object.read()
    
parser = MyHTMLParser()
parser.feed(contents)

for i in range(len(parser.totalList)):
    if 'shape' in str(parser.totalList[i][1]) and len(parser.totalList[i][1])==8 and len(parser.totalList[i])>2:
        print(parser.totalList[i])
```
