```python
# 재귀호출 테스트 #1 - 조건문을 뒤쪽에 배치

point='6317'
matches='GMSKKMGSKGMS'
match_result='LLLLLL'

def determine(point,teams='',others=matches):
    max_value = max(point)
    new_point=''
    for t,p in zip(others,point):
        if p == max_value:
            teams+=t
        else :
            others+=t
            new_point+=p
    if len(teams)<2:
        teams = determine(new_point,teams,others)
    return teams

print(determine(point))
```
